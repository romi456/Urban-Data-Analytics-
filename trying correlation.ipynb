```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('~/this_file/secondCopy.csv')
df.head()  # View the first few rows of the dataset
```

```python
correlation_matrix = df[['percent_less_than_LIM', 'pm25_2016', 'greenness_2016']].corr()
correlation_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix for 2016 NDVI values, 2016 below LIM percentage, and 2016 PM2.5 concentration')
plt.show()
```

```python
sns.scatterplot(data=df, x='greenness_2016', y='percent_less_than_LIM')
plt.title('Scatter Plot of NDVI vs. LIM')
plt.show()
```

```python
sns.scatterplot(data=df, x='greenness_2016', y='pm25_2016')
plt.title('Scatter Plot of NDVI vs. LIM')
plt.show()
```
